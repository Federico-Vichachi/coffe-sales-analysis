```python
import pandas as pd
```

```python
df1 = pd.read_csv("../data/raw/index_1.csv")
df2 = pd.read_csv("../data/raw/index_2.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.info()
df2.info()
```

```python
df1.drop(['card'], axis=1, inplace=True)
df1
```

```python
df1['datetime'] = pd.to_datetime(df1['datetime']).dt.strftime('%H:%M:%S')
df1
```

```python
df2['datetime'] = pd.to_datetime(df1['datetime']).dt.strftime('%H:%M:%S')
df2
```

```python
df = pd.concat([df1, df2])
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
df.rename(columns={'datetime' : 'hour', 'cash_type' : 'payment_method'}, inplace=True)
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df['hour'] = pd.to_datetime(df['hour'])
df.info()
```

```python
df['coffee_name'].value_counts()
```

```python
caffeinated_keywords = ['americano', 'latte', 'capuccino', 'cortado', 'espresso', 'coffe', 'double', 'mochaccino', 'with']
df['has_caffeine'] = df['coffee_name'].str.lower().apply(
    lambda x: any(keyword in x for keyword in caffeinated_keywords) 
)
df['has_caffeine'].value_counts()
```

```python
milk_drinks_keywords = ['milk', 'latte', 'cappuccino', 'cortado', 'chocolate', 'cocoa', 'espresso', 'mochaccino']
df['has_milk'] = df['coffee_name'].str.lower().apply(
    lambda x: any(keyword in x for keyword in milk_drinks_keywords)
)
df
```

```python

alcoholic_drinks_keywords = ['irish', 'whiskey']
df['has_alcohol'] = df['coffee_name'].str.lower().apply(
    lambda x: any(keyword in x for keyword in alcoholic_drinks_keywords)
)
df
```

```python
df.to_csv('../output/clean_data.csv')
```
